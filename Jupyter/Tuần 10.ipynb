```python
#1
def merge_sort(arr):
    if len(arr)>1:
        mid=len(arr)//2
        left_half=arr[:mid]
        right_half=arr[mid:]
        #recursion
        merge_sort(left_half)
        merge_sort(right_half)
        # main logic
        i=j=k=0
        while i<len(left_half) and j <len(right_half):
            if left_half[i]<right_half[j]:    
                arr[k]=left_half[i]
                i+=1
            else:
                arr[k]=right_half[j]
                j+=1
            k+=1
        while i<len(left_half):
            arr[k]=left_half[i]
            i+=1
            k+=1
        while j<len(right_half):
            arr[k]=right_half[j]
            j+=1
            k+=1
ds1=list(map(int,input().split()))
ds2=list(map(int,input().split()))
dsm=ds1+ds2
merge_sort(dsm)
print(dsm) 
```

```python
#2
def count_less_equal(matrix,n,m,target):
    row=0
    col=m-1
    count=0
    while row<n and col>=0:
        if matrix[row][col]<=target:
            count+=(col+1)
            row+=1
        else:
            col -=1
    return count
def find(matrix, k,n,m):
    low=matrix[0][0]
    high=matrix[n-1][m-1]
    
    while low<high:
        mid=(low+high)//2
        count=count_less_equal(matrix,n,m,mid)
        if count<k:
            low=mid+1 
        else:
            high=mid
    return low
            


```
