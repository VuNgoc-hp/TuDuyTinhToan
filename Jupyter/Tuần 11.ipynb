```python
#3
try:
    a=int(input())
    seen=set()
    newa=0
    while True:
        while a>0:
            newa+=(a%10)**2
            a=a//10
        if newa in seen:
            print("NO")
            break
        if newa==1:
            print("YES")
            break
        seen.add(newa)
        a=newa
        newa=0
except ValueError:
    print("Loi")
        
```
