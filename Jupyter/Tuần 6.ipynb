```python
#10
lst=list(map(int,input().split()))
k=int(input())
def tong(ds,n):
    seen_value=set()
    result=set()
    for i in ds:
        target=n-i
        if target in seen_value:
            result.add((target,i))
        seen_value.add(i)
    sorter_result=sorted(list(result))
    print(sorter_result)
tong(lst,k)


```

```python
#12
def find_largest_frequency(lst):
    dic={}
    for i in lst:
        if i not in dic:
            dic[i]=1
        else:
            dic[i]+=1
    maxx_freq=float('-inf')
    res=float('inf')
    for key,value in dic.items():   
        if value>maxx_freq:
            maxx_freq=value
            res=key
        elif value==maxx_freq:
            if key<res:
                res=key
    return res
lst=[int(i) for i in input().split()]
print(find_largest_frequency(lst))


```
