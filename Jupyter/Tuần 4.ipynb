```python
#W4A6
a,b=map(int,input().split())
def sangnt(a,n):
    sang=[1]*(n+1)
    sang[0]=sang[1]=0
    for i in range (2,int(n*0.5)+1):
        if sang[i]==1:
            for j in range (i*i,n+1,i):
                sang[j]=0
    tong=0
    for i in range (a,n+1):
        if sang[i]==1:
            tong+=i
    return tong
result=sangnt(a,b)
print(result)
```

```python
#W4A8
def check_palindrome(st):
    if st==st[::-1]:
        return True
    else:
        return False
n=input()
count=0
while check_palindrome(n) == False:
    n=str(int(n[::-1])+int(n))
    count+=1

print(count,n)
```

```python
#W4A9
def pairwise_distinct(n):
    check=[0]*10
    for i in str(n):
        if check[int(i)]==1:
            return False
        else:
            check[int(i)]+=1
    return True


n=50
result=[]
for i in range (1,int(n**(0.5))+1):
    if pairwise_distinct(i**2):
        result.append(i**2)
if len(result)==0:
    print("no number")
else:
    print(*result)
```
