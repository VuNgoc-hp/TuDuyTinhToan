```python
#1
class cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def S_be_mat(self):
        return self.h*2*3.14*self.r+2*3.14*self.r**2
    def the_tich(self):
        return self.h*3.14*self.r**2
a,b=map(float,input().split())
ktra=cylinder(a,b)
dientich=ktra.S_be_mat()
thetich=ktra.the_tich()
print(dientich, thetich)
```

```python
#2
class Calendar:
    def __init__(self,ngay,thang,nam):
        self.ngay=ngay
        self.thang=thang
        self.nam=nam
    def translate(self):
        if self.nam%4==0:
            if self.thang in [1,3,5,7,8,10,12]:
                if self.ngay<=31:
                    return True
                return False
            elif self.thang==2:
                if self.ngay<=29:
                    return True
                return False
            elif self.thang in [4,6,9,11]:
                if self.ngay<=30:
                    return True
                return False
        else:
            if self.thang in [1,3,5,7,8,10,12]:
                if self.ngay<=31:
                    return True
                return False
            elif self.thang==2:
                if self.ngay<=28:
                    return True
                return False
            elif self.thang in [4,6,9,11]:
                if self.ngay<=30:
                    return True
                return False
```
