```python
#1
lst=list(map(int,input().split()))
k=int(input())
left=0
right=len(lst)-1
kq=-1
while left<=right:
    mid=(left+right)//2
    if lst[mid]==k:
        kq=mid
        break
    elif lst[mid]>k:
        right=mid-1
    else:
        left=mid+1
print(kq)
```

```python
#2
lst=list(map(int,input().split()))
x=int(input())
def count_occurrences(lst, x):
    dic={}
    for i in lst:
        if x ==i:
            dic[x]=dic.get(x,0)+1
    return dic.get(x,0)
print(count_occurrences(lst,x))
```

```python
#3
arr=list(map(int,input().split()))
def bubble_sort(arr):
    x=len(arr)
    for i in range (x):
        swaped=False
        for j in range (0, x-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swaped=True
        if swaped==False:
            break
    return arr
def selection_sort(arr):
    x=len(arr)
    for i in range (x-1):
        smallest_num=float("inf")
        position=0
        for j in range (i,x):
            if arr[j]<=smallest_num:
                smallest_num=arr[j]
                position=j
        arr[i],arr[position]=smallest_num,arr[i]
    return arr
print(bubble_sort(arr))
print(selection_sort(arr))

        
```

```python
#4
arr=list(map(int,input().split()))
def find_most_occurrences(arr):
    dic={}
    for i in arr:
        dic[i]=dic.get(i,0)+1
    largest_dependency=float('-inf')
    res=0
    for key, val in dic.items():
        if val>largest_dependency:
            largest_dependency=val
            res=key
    return res
print(find_most_occurrences(arr))
```

```python
#5
arr=list(map(int,input().split()))
x=int(input())
def find_couple(arr,x):
    seen={}
    count=0
    for i in arr:
        target=x-i
        if target in seen:
            count+=seen[target]
        if i in seen:
            seen[i]+=1
        else:
            seen[i]=1
    return count
print(find_couple(arr,x))
```
